```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la señal
t = np.linspace(0, 1, 500)  # Tiempo con 500 muestras
f0 = 5                      # Frecuencia de la señal
señal = np.sin(2 * np.pi * f0 * t)  # Señal senoidal

# Muestras cada intervalo para la primera visualización
intervalo = 10
puntos_t = t[::intervalo]  # Tiempo en los puntos seleccionados
puntos_señal = señal[::intervalo]  # Valores de la señal en los puntos seleccionados

# Graficar la señal y los puntos seleccionados
plt.figure(figsize=(10, 4))
plt.plot(t, señal, label="Señal Senoidal", color='b')
plt.scatter(puntos_t, puntos_señal, color='red', marker='o', label="Puntos Marcados")
plt.title("Señal Senoidal con Puntos Marcados")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.legend()
plt.grid(True)
plt.show()

# Visualización con más puntos
t1 = np.linspace(0, 1, 1000)  # Mayor resolución temporal
señal1 = np.sin(2 * np.pi * f0 * t1)

intervalo1 = 1
puntos_t1 = t1[::intervalo1]  # Tiempo en los puntos seleccionados
puntos_señal1 = señal1[::intervalo1]  # Valores de la señal en los puntos seleccionados

# Graficar la señal con mayor resolución temporal
plt.figure(figsize=(10, 4))
plt.plot(t1, señal1, label="Señal Senoidal", color='b')
plt.scatter(puntos_t1, puntos_señal1, color='red', marker='o', label="Puntos Marcados")
plt.title("Señal Senoidal con Puntos Marcados")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.legend()
plt.grid(True)
plt.show()

```

```python
# Análisis en el dominio de la frecuencia usando FFT
# Calculamos la FFT y la frecuencia correspondiente para la primera señal
fft_señal = np.fft.fft(señal)
frecuencia = np.fft.fftfreq(len(t), d=(t[1] - t[0]))
#fft_señal = np.fft.fftshift(fft_señal)  # Centramos la FFT
#frecuencia = np.fft.fftshift(frecuencia)



# Para la señal con más puntos
fft_señal1 = np.fft.fft(señal1)
frecuencia1 = np.fft.fftfreq(len(t1), d=(t1[1] - t1[0]))
#fft_señal1 = np.fft.fftshift(fft_señal1)  # Centramos la FFT
#frecuencia1 = np.fft.fftshift(frecuencia1)

```

```python
# Espectro de frecuencia de la señal con más puntos
plt.figure(figsize=(10, 4))
plt.plot(frecuencia, fft_señal, color='purple')
plt.title("Espectro de Frecuencia de la Señal Senoidal")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()

# Espectro de frecuencia de la señal con más puntos
plt.figure(figsize=(10, 4))
plt.plot(frecuencia1, fft_señal1, color='purple')
plt.title("Espectro de Frecuencia de la Señal Senoidal con Mayor Resolución")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()

```

```python
fft_señal = np.fft.fftshift(fft_señal)  # Centramos la FFT
fft_señal1 = np.fft.fftshift(fft_señal1)  # Centramos la FFT

# Espectro de frecuencia de la señal con más puntos
plt.figure(figsize=(10, 4))
plt.plot(frecuencia, fft_señal, color='purple')
plt.title("Espectro de Frecuencia de la Señal Senoidal")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()

# Espectro de frecuencia de la señal con más puntos
plt.figure(figsize=(10, 4))
plt.plot(frecuencia1, np.abs(fft_señal1), color='purple')
plt.title("Espectro de Frecuencia de la Señal Senoidal con Mayor Resolución")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()
```

```python
frecuencia = np.fft.fftshift(frecuencia)
frecuencia1 = np.fft.fftshift(frecuencia1)

# Espectro de frecuencia de la señal con más puntos
plt.figure(figsize=(10, 4))
plt.plot(frecuencia, np.abs(fft_señal), color='purple')
plt.title("Espectro de Frecuencia de la Señal Senoidal")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()

# Espectro de frecuencia de la señal con más puntos
plt.figure(figsize=(10, 4))
plt.plot(frecuencia1, np.abs(fft_señal1), color='purple')
plt.title("Espectro de Frecuencia de la Señal Senoidal con Mayor Resolución")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()
```
